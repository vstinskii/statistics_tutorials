### Venn Diagrams
<pre>
If P(A) - probability of event A: then P(A) + P(not A) = 1
If A and B are events then:
    P(A or B) = P(A) + P(B) - P(A and B) 
    # - P(A and B) - because of doubled intersection

For mutually exclusive events:
    P(A and B) = 0
so  P(A or B) = P(A) + P(B) [*] Why there is a plus here?

Notation of Conditional Probability:
    P(A|B) = probability of A given that B has occured
    P(A|B) = P(A and B) / P(B) (if P(B) is not zero)

If A & B are independent form each other, then:
    P(A|B) = P(A)
    P(B|A) = P(B)
